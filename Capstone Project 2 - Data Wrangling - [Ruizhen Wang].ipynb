```python
#the StandardHours has a universal value: 80 hours, so we could drop this column
HRdf.drop(columns=['StandardHours'], inplace=True)
```

```python
#the EmployeeCount also have a universal value: 1, so we also drop this column
HRdf.drop(columns=['EmployeeCount'], inplace=True)
```

```python
# there are DailyRate, HourlyRate, MonthlyRate which may relate to the employees' salary, while MonthlyIncome should be relate to employees' salary. 
# But it is not very clear how to interpret the three rate columns without clear guidance from original dataset. 
# Let's do some visulization for these four columns.
# First, let's do pairplot to explore relationships between all pairs of these columns simultaneously,
# visualize correlations and patterns between all combinations of the columns.
sns.pairplot(HRdf[['DailyRate', 'HourlyRate', 'MonthlyRate', 'MonthlyIncome']])
plt.show()
```

```python
# we could see from the pairplot that three Rate columns has poor relationship with each other, and with MontlyIncome
# Also, three Rate columns all have uniform distribution, which all values have roughly the same frequency, and there are no significant peaks or troughs in the data.
# Only MonthlyIncome displays a right-skewed distribution. 
# We also do a correlation heatmap to verify the correlation between these four columns. 
correlation_matrix = HRdf[['DailyRate', 'HourlyRate', 'MonthlyRate', 'MonthlyIncome']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
```

```python
# We could say three Rate columns has very poor relationship with each other and with MonthlyIncome. 
# Plus, if the hourly rate were 'calculated' for salary employees, this calculations is useless without the number of hours the employee is actually working.
# So I find hourly, daily, and monthly rate to be poor variables for this analysis.
# So I drop the hourly, daily, and monthly rates, and focused on the monthly income instead.
HRdf.drop(columns=['DailyRate', 'HourlyRate', 'MonthlyRate'], inplace=True)
```

```python
# Convert categorical columns to numerical, for further data analysis and machine learning.
#'Yes' to 1 and 'No' to 0 for Attrition
#'Yes' to 1 and 'No' to 0 for OverTime
#'Male' to 1 and 'Female' to 0 for Gender
mapping = {'Yes': 1, 'No': 0}

HRdf['Attrition'] = HRdf['Attrition'].replace(mapping)
HRdf['OverTime'] = HRdf['OverTime'].replace(mapping)

gender_mapping = {'Male': 1, 'Female': 0}
HRdf['Gender'] = HRdf['Gender'].replace(gender_mapping)
```

```python
# As BusinessTravel has an inherent order, so convert BusinessTravel to numercial
BusinessTravel_mapping = {'Travel_Rarely': 2, 'Travel_Frequently': 3, 'Non-Travel': 1}
HRdf['BusinessTravel'] = HRdf['BusinessTravel'].map(BusinessTravel_mapping)
```
